```python
df.rename(columns = {'Краткое\nсодержание':'Содержание'}, inplace = True)
df
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

content=[]
from nltk.tokenize import sent_tokenize, word_tokenize
text = df.loc[:,'Содержание']
print(text)
for i in text:
    content.append(word_tokenize(i, language="russian"))
print(content)
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = nltk.corpus.stopwords.words('russian')
stopwords.append('вх')
stopwords.append('об')
stopwords.append('-')
stopwords.append('"')
stopwords.append('``')
stopwords.append(',')
stopwords.append('.')
stopwords.append('»')
stopwords.append('«')
stopwords.append('(')
stopwords.append('\'\'')
stopwords.append(')')
stopwords
```

```python
new_content=content
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language="russian")
for x in range(0,len(new_content)):
    for y in range(0,len(new_content[x])):
        ss = new_content[x][y]
        new_content[x][y]=snowball.stem(ss)
new_content
```

```python
stop_words = stopwords
filtered_str=[]
new_content2=[]
for x in range(0,len(new_content)):
    for y in range(0,len(new_content[x])):
        filtered_str=' '.join((filter(lambda s: s not in stop_words,new_content[x])))
    new_content2.append(filtered_str)
new_content2
```

```python
content = pd.Series(new_content2)
content
```

```python
content=pd.DataFrame({'Содержание':content})
content
```

```python
df=df.drop(['Содержание'], axis = 1)
df
```

```python
df.insert(4, "Содержание", content, True)
df
```
