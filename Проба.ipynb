```python
df=pd.read_csv('Новая выгрузка обр.csv', encoding="cp1251")
df
```

```python
new_row = {'Автор': ' ', 'Место работы автора':' ', 'Адресат':' ', 'Содержание':'участ чемпионат WorldSkills Hi-Tech екатеринбу...', 'Примечание': ' ', 'Специализация адресата': ' '}
df = df.append(new_row, ignore_index=True)
```

```python
df
```

```python
X = df['Содержание']
y = df["Специализация адресата"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=1500, min_df=5, max_df=0.7, stop_words=stopwords.words('russian'))
X = vectorizer.fit_transform(X).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidfconverter = TfidfTransformer()
X = tfidfconverter.fit_transform(X).toarray()
```

```python
X_test = X[-1]
X_test
```

```python
X_test = X_test.reshape(1, -1)
```
