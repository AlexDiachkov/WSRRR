```python
autor=df_new.loc[:,'Автор']
autor
```

```python
work=[]
emp=[]
for i in autor:
    work.append(i.split('(')[1].strip())
    emp.append(i.split('(')[0].strip())
work
```

```python
work2=[]
for i in work:
    x=i.replace(')', "")
    work2.append(x)

work2
```
